3. Navigating the parse tree and iterate through tags: once the soup is made, we have the HTML code in Python. We can then find our desired information by searching through HTML tags.

```python
import json
import pandas as pd
from bs4 import BeautifulSoup

# Sample HTML content provided in the question
html_content = """
    <!-- Your HTML content here -->
"""

# Create a BeautifulSoup object
soup = BeautifulSoup(html_content, 'html.parser')

# Initialize a list to store extracted data
extracted_data = []

# Initialize a set to track unique ZPID values
unique_zpids = set()

# Find all script tags with type application/ld+json
json_data_tags = soup.find_all("script", {"type": "application/ld+json"})

# Loop through each script tag
for json_data_tag in json_data_tags:
    # Extract content from the script tag
    json_data_content = json_data_tag.string.strip()
    
    # Skip empty script content
    if not json_data_content:
        continue
    
    # Parse the script content as JSON
    data = json.loads(json_data_content)

    # Check if the data is a dictionary
    if isinstance(data, dict):
        # Extract relevant information
        url = data.get('url', 'N/A')
        zpid = url.split('/')[-2].split('_')[0] if url else 'N/A'

        # Check if the ZPID is unique
        if zpid not in unique_zpids:
            # Add ZPID to the set of unique ZPIDs
            unique_zpids.add(zpid)

            # Extract address, price, and other relevant information
            address = data.get('address', {}).get('streetAddress', 'N/A')
            locality = data.get('address', {}).get('addressLocality', 'N/A')
            region = data.get('address', {}).get('addressRegion', 'N/A')
            postal_code = data.get('address', {}).get('postalCode', 'N/A')
            floor_size = data.get('floorSize', {}).get('value', 'N/A')
            price = data.get('offers', {}).get('price', 'N/A')

            # Append extracted data to the list
            extracted_data.append({
                'ZPID': zpid,
                'Price, $': price,
                'Address': address,
                'Locality': locality,
                'Region': region,
                'Postal Code': postal_code,
                'Floor Size': floor_size,
                'URL': url
            })

# Create a DataFrame from the extracted data
df = pd.DataFrame(extracted_data)

# Print the DataFrame
print(df)


#df.index = np.arange(1, len(df) + 1)
#df.drop(df.tail(1).index,inplace=True) # drop last n rows
```
